```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
df = pd.read_csv('pldb.csv')
```

```python
print('Dataset Shape:', df.shape)
```

```python
print(df.describe(include='all'))
```

```python
print(df['type'].unique())
print(df['type'].value_counts())
```

```python
pl_only = df[df['type'] == 'pl']
oldest_year = pl_only['appeared'].min()
print(pl_only[pl_only['appeared'] == oldest_year][['title', 'appeared']])
```

```python
appeared_2022 = pl_only[pl_only['appeared'] == 2022]
appeared_2022[['title', 'appeared', 'type']]
```

```python
top_users = pl_only[['title', 'numberOfUsers']].dropna().sort_values(by='numberOfUsers', ascending=False).head(10)
print(top_users)
```

```python
top_jobs = pl_only[['title', 'numberOfJobs']].dropna().sort_values(by='numberOfJobs', ascending=False).head(10)
print(top_jobs)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_users['title'], top_users['numberOfUsers'])
plt.xticks(rotation=45)
plt.title('Top 10 Programming Languages by Users')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_jobs['title'], top_jobs['numberOfJobs'])
plt.xticks(rotation=45)
plt.title('Top 10 Programming Languages by Jobs')
plt.tight_layout()
plt.show()
```

```python
top_books = pl_only[['title', 'bookCount']].dropna().sort_values(by='bookCount', ascending=False).head(10)
print(top_books)
```

```python
plt.figure(figsize=(8,8))
plt.pie(top_books['bookCount'], labels=top_books['title'], autopct='%1.1f%%')
plt.title('Top 10 Programming Languages by Book Count')
plt.tight_layout()
plt.show()
```
