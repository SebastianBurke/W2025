### Data Analysis Using Pandas

```python
import pandas as pd
df = pd.read_csv('global_unemployment.csv')
df.head()
```

Dropping rows (0, 1, 2)

```python
df.drop([0,1,2], axis=0, inplace=True)
df.head()
```

Make first row the column header and drop the row

```python
df.columns = df.iloc[0]
df.drop(labels=3, axis=0, inplace=True)
df.head()
```

Check # of rows and columns

```python
df.shape
```

Check columns

```python
df.columns
```

New dataframe five_year_df

```python
five_year_df = df[["Country Name", "Country Code", 2017.0, 2018.0, 2019.0, 2020.0, 2021.0]]
five_year_df.head(7)
five_year_df.tail(7)
```

Checking for null values using isnull() and sum()

```python
five_year_df.isnull().sum()
```

Fill in null values using ffill()

```python
new_five_year_df = five_year_df.ffill()
new_five_year_df.isnull().sum()
```

```python
new_five_year_df.head(7)
```

Drop row for Aruba which is completely null

```python
new_five_year_df = new_five_year_df.dropna()
new_five_year_df.head(7)
```

```python
new_five_year_df.isnull().sum()
```

Check for duplicates

```python
new_five_year_df.duplicated().sum()
```

Check dtypes

```python
new_five_year_df.dtypes
```

Sort by Country Name

```python
new_five_year_df = new_five_year_df.sort_values("Country Name")
new_five_year_df.head()
```

Set index of dataframe to Country Code

```python
new_five_year_df.set_index(["Country Code"], inplace=True)
new_five_year_df.head()
```

Add a new column called 5yrs_avg

```python
new_five_year_df["5yrs_avg"] = new_five_year_df[[2017.0, 2018.0, 2019.0, 2020.0, 2021.0]].mean(axis=1)
new_five_year_df.head()
```

Add a new column called covid_yr_dff

```python
new_five_year_df["covid_yr_diff"] = new_five_year_df[2020.0] - new_five_year_df[2019.0]
new_five_year_df.head()
```

List all countries with unemployment greater than or equal to 25% in 2021

```python
high_unemployment_2021 = new_five_year_df[new_five_year_df[2021.0] >= 25]
high_unemployment_2021[["Country Name", 2021.0]]
```

List all the countries where unemployment is less than or equal to 1% in 2021

```python
low_unemployment_2021 = new_five_year_df[new_five_year_df[2021.0] <= 1]
low_unemployment_2021[["Country Name", 2021.0]]
```

List the top 10 countries that had an average high rate of unemployment from 2017 to 2021

```python
top_10_avg_high = new_five_year_df.sort_values("5yrs_avg", ascending=False).head(10)
top_10_avg_high[["Country Name", "5yrs_avg"]]
```

List the top 10 countries that had an average low rate of unemployment from 2017 to 2021

```python
top_10_avg_high = new_five_year_df.sort_values("5yrs_avg", ascending=False).head(10)
top_10_avg_high[["Country Name", "5yrs_avg"]]
```

List the top 10 countries with a high COVID impact on employment rate

```python
top_10_covid_impact = new_five_year_df.sort_values("covid_yr_diff", ascending=False).head(10)
top_10_covid_impact[["Country Name", "covid_yr_diff"]]
```

List the top 10 countries that had a low COVID impact on employment rate

```python
top_10_covid_least_impact = new_five_year_df.sort_values("covid_yr_diff", ascending=True).head(10)
top_10_covid_least_impact[["Country Name", "covid_yr_diff"]]
```
