```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("data_files/fish.csv")
df.head()
```

```python
print(df.info())
print(df.describe())
print(df.head())
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
axs[0].scatter(df['Length1'], df['Weight'])
axs[0].set_title('Length1 vs Weight')
axs[0].set_xlabel('Length1')
axs[0].set_ylabel('Weight')

axs[1].scatter(df['Length2'], df['Weight'])
axs[1].set_title('Length2 vs Weight')
axs[1].set_xlabel('Length2')
axs[1].set_ylabel('Weight')

axs[2].scatter(df['Length3'], df['Weight'])
axs[2].set_title('Length3 vs Weight')
axs[2].set_xlabel('Length3')
axs[2].set_ylabel('Weight')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df['Weight'], bins=20, edgecolor='black')
plt.title('Distribution of Weight')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```

```python
df_clean = df.drop('Species', axis=1)
y = df_clean['Weight']
x = df_clean.drop('Weight', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1234)
model = LinearRegression()
model.fit(x_train, y_train)

intercept = model.intercept_
coefficients = model.coef_
score = model.score(x_test, y_test)
y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, y_pred)

(intercept, coefficients, score, mae)
```
